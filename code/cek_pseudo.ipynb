# Gabungkan labels.csv dan pseudolabelling_ver_1.csv

Menggabungkan kedua file berdasarkan File_Name yang sama, dengan menambahkan suffix "ver 1" dan "ver 2" pada nama kolom

```python
import pandas as pd

# Baca kedua file
labels = pd.read_csv('../Dataset/labels.csv')
pseudo = pd.read_csv('../Dataset/pseudolabelling_ver_1.csv')

print(f"labels.csv: {labels.shape}")
print(f"pseudolabelling_ver_1.csv: {pseudo.shape}")
```

```python
# Cek file yang ada di kedua dataset
file_labels = set(labels['File_Name'].unique())
file_pseudo = set(pseudo['File_Name'].unique())

file_sama = file_labels.intersection(file_pseudo)
file_hanya_labels = file_labels - file_pseudo
file_hanya_pseudo = file_pseudo - file_labels

print(f"üìÅ File di labels.csv: {len(file_labels)}")
print(f"üìÅ File di pseudolabelling_ver_1.csv: {len(file_pseudo)}")
print(f"‚úÖ File yang SAMA di kedua dataset: {len(file_sama)}")
print(f"‚ö†Ô∏è File hanya di labels.csv: {len(file_hanya_labels)}")
print(f"‚ö†Ô∏è File hanya di pseudolabelling_ver_1.csv: {len(file_hanya_pseudo)}")
```

```python
# Tambahkan suffix "ver 1" dan "ver 2" pada nama kolom (kecuali File_Name yang akan digunakan untuk merge)
labels_renamed = labels.copy()
pseudo_renamed = pseudo.copy()

# Rename semua kolom kecuali File_Name
for col in labels.columns:
    if col != 'File_Name':
        labels_renamed.rename(columns={col: f"{col} ver 1"}, inplace=True)

for col in pseudo.columns:
    if col != 'File_Name':
        pseudo_renamed.rename(columns={col: f"{col} ver 2"}, inplace=True)

print("Kolom labels (ver 1):", labels_renamed.columns.tolist())
print("Kolom pseudo (ver 2):", pseudo_renamed.columns.tolist())
```

```python
# Gabungkan berdasarkan File_Name (inner join - hanya yang ada di kedua file)
gabungan = pd.merge(labels_renamed, pseudo_renamed, on='File_Name', how='inner')

print(f"‚úÖ Berhasil menggabungkan!")
print(f"üìä Total baris setelah penggabungan: {len(gabungan)}")
print(f"üìä Total kolom: {len(gabungan.columns)}")
print(f"\nKolom gabungan:")
for i, col in enumerate(gabungan.columns, 1):
    print(f"{i}. {col}")
```

## Analisis Data

```python
# Simpan ke file CSV baru
gabungan.to_csv('../Dataset/gabungan.csv', index=False)
print("‚úÖ File berhasil disimpan ke: Dataset/gabungan.csv")
```

```python
# Cek jumlah total baris
total_baris = len(gabungan)
print(f"üìä Total jumlah baris yang digabungkan: {total_baris}")
```

```python
# Bandingkan label ver 1 dan ver 2
label_berbeda = gabungan['Label ver 1'] != gabungan['Label ver 2']
jumlah_berbeda = label_berbeda.sum()

print(f"\nüîç Jumlah baris dengan label BERBEDA: {jumlah_berbeda}")
print(f"‚úÖ Jumlah baris dengan label SAMA: {total_baris - jumlah_berbeda}")
print(f"üìà Persentase perbedaan: {(jumlah_berbeda/total_baris)*100:.2f}%")
```

```python
# Tampilkan data yang labelnya berbeda
if jumlah_berbeda > 0:
    data_berbeda = gabungan[label_berbeda][['File_Name', 'Label ver 1', 'Label ver 2', 'Teks_Terlihat ver 1']]
    print(f"\nüìã Contoh {min(10, jumlah_berbeda)} baris dengan label berbeda:")
    display(data_berbeda.head(10))
else:
    print("\n‚úÖ Tidak ada perbedaan label!")
```

```python
# Breakdown perbedaan label
if jumlah_berbeda > 0:
    print("\nüîÑ Breakdown perubahan label:")
    perubahan = gabungan[label_berbeda].groupby(['Label ver 1', 'Label ver 2']).size().reset_index(name='Jumlah')
    perubahan = perubahan.sort_values('Jumlah', ascending=False)
    display(perubahan)
else:
    print("\n‚úÖ Semua label sama, tidak ada perubahan!")
```
